```python
# За секој град пресметајте колку ресторани има во него.
cities = df['city'].unique()
print(cities)
for city in cities:
  count = df[df['city']==city]['name'].nunique()
  print('There are', count,'restaurants in', city)
```

```python
# Кој ресторан има најмал број на нарачки во секој град?

unique_cities = df['city'].unique().tolist()

for c in unique_cities:
    restaurant = df[df['city'] == c]['name'].value_counts(ascending = True).index[0]
    print('The restaurant with the least amount of orders in', c, "is", restaurant)
```

```python
# Пресметајте ги рестораните со највисоки оценки.
# Hint: користете max() за да ја пронајдете највисоката оценка

df[df['aggregate_rating'] == df['aggregate_rating'].max()]
```
